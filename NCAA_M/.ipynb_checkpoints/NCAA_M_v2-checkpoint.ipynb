### 1. 출처 : [Google Cloud & NCAA® ML Competition 2020-NCAAM](https://www.kaggle.com/c/google-cloud-ncaa-march-madness-2020-division-1-mens-tournament/overview)


### 2. 목표 : NCAA 남자팀 과거 경기를 토대로 Tournament 경기 결과 예측(Stage 1)


### 3. Column 설명

### MTeams

- TeamID : 팀 ID
- TeamName : 팀 이름
- FirstD1Season - 첫 1부 리그 진입 시즌
- LastD1Season - 마지막 1부 리그 진입 시즌

### MSeasons

- Season : 시즌
- DayZero : 시즌 시작일
- RegionW, RegionX, Region Y, Region Z : 리그 지역

### MNCAATourneySeeds

- Season : 시즌
- Seed : 토너먼트 시드 번호 
- TeamID : 팀 ID

### MRegularSeasonCompactResults, MNCAATourneyCompactResults

- Season : 시즌 
- DayNum : 경기 번호 
- WTeamID : 승리팀 ID
- WScore : 승리팀 득점
- LTeamID : 패배팀 ID
- LScore : 패배팀 득점
- WLoc : 승리 경기장(H : 홈 경기장, A : 원정 경기장, N : 중립 경기장)
- NumOT : 연장전 횟수

### MSampleSubmissionStage1

- ID : ID, SSSS_XXXX_YYYY 형태로 되어 있음(시즌_승리팀 ID_패배팀 ID)
- Pred : 승리팀 승률 예측

# Data Merge

## Create Regular Season Data

```python
merge1 = MRegularSeasonCompactResults.merge(Mteams,
                                            left_on = "WTeamID",
                                            right_on = "TeamID",
                                            how = "left")

merge1 = merge1.drop("TeamID", axis = 1)

merge1 = merge1.rename(columns = {"TeamName" : "WTeamName"}) 

regular = merge1.merge(Mteams,
                      left_on = "LTeamID",
                      right_on = "TeamID",
                      how = "left")

regular = regular.drop("TeamID", axis = 1)

regular = regular.rename(columns = {"TeamName" : "LTeamName"}) 

print(regular.shape)
regular.head()
```

```python
columns = ["Season", "WTeamID", "WTeamName", "WScore",
           "LTeamID", "LTeamName", "LScore", "WLoc", "NumOT"]

regular = regular[columns]

print(regular.shape)
regular.head()
```

## Create Tournament Season Data

```python
merge1 = MNCAATourneyCompactResults.merge(MNCAATourneySeeds,
                                          left_on = ["Season", "WTeamID"],
                                          right_on = ["Season", "TeamID"],
                                          how = "left")

merge1 = merge1.drop("TeamID", axis = 1)

tournament = merge1.merge(MNCAATourneySeeds,
                          left_on = ["Season", "LTeamID"],
                          right_on = ["Season", "TeamID"],
                          how = "left")

tournament = tournament.drop(["DayNum", "TeamID"], axis = 1)

print(tournament.shape)
tournament.head()
```

```python
tournament = tournament.merge(Mteams,
                              left_on = "WTeamID",
                              right_on = "TeamID",
                              how = "left")

tournament = tournament.drop("TeamID", axis = 1)

tournament = tournament.merge(Mteams,
                              left_on = "LTeamID",
                              right_on = "TeamID",
                              how = "left")

tournament = tournament.drop("TeamID", axis = 1)

print(tournament.shape)
tournament.head()
```

```python
drop_columns = ["FirstD1Season_x", "LastD1Season_x", "FirstD1Season_y", "LastD1Season_y"]

tournament = tournament.drop(drop_columns, axis = 1)

print(tournament.shape)
tournament.head()
```

```python
order_columns = ['Season', 'WTeamID', "TeamName_x",'WScore', 'LTeamID', "TeamName_y", 'LScore',
                 'WLoc', 'NumOT', 'Seed_x', 'Seed_y']

tournament = tournament[order_columns]

print(tournament.shape)
tournament.head()
```

```python
new_columns = ["Season", "WTeamID_T", "WTeamName_T", "WScore_T", "LTeamID_T", "LTeamName_T", "LScore_T",
               "WLoc_T", "NumOT_T", "WSeed", "LSeed"]

tournament.columns = new_columns

print(tournament.shape)
tournament.head()
```

# Create Train Data

정규 시즌의 시즌별 / 팀별 점수를 합하여 별도의 데이터 프레임으로 만들겠습니다.

```python
regular_win = regular[["Season", "WTeamID", "WScore"]]
regular_win.columns = ["Season", "TeamID", "Score"]

regular_lose = regular[["Season", "LTeamID", "LScore"]]
regular_lose.columns = ["Season", "TeamID", "Score"]

regular_score = pd.concat([regular_win, regular_lose])

print(regular_score.shape)
regular_score.head()
```

```python
regular_score = regular_score.groupby(["Season", "TeamID"])["Score"].sum().reset_index()

print(regular_score.shape)
regular_score.head()
```

토너먼트의 시즌별 / 팀별 점수를 합하여 별도의 데이터 프레임으로 만들겠습니다.

```python
tournament_win = tournament[["Season", "WTeamID_T", "WScore_T"]]
tournament_win.columns = ["Season", "TeamID", "Score_T"]

tournament_lose = tournament[["Season", "LTeamID_T", "LScore_T"]]
tournament_lose.columns = ["Season", "TeamID", "Score_T"]

tournament_score = pd.concat([tournament_win, tournament_lose])

print(tournament_score.shape)
tournament_score.head()
```

```python
tournament_score = tournament_score.groupby(["Season", "TeamID"])["Score_T"].sum().reset_index()

print(tournament_score.shape)
tournament_score.head()
```

토너먼트 결과에 정규 시즌별 / 팀별 점수 합계를 합치겠습니다.

```python
tournament_result = tournament[["Season", "WTeamID_T", "LTeamID_T", "WSeed_num", "LSeed_num"]]

print(tournament_result.shape)
tournament_result.head()
```

```python
tournament_result = pd.merge(tournament_result, regular_score,
                             left_on = ["Season", "WTeamID_T"],
                             right_on = ["Season", "TeamID"],
                             how = "left")

tournament_result = tournament_result.rename(columns = {"Score" : "WScore"})
tournament_result = tournament_result.drop("TeamID", axis = 1)

print(tournament_result.shape)
tournament_result.head()
```

```python
tournament_result = pd.merge(tournament_result, regular_score,
                             left_on = ["Season", "LTeamID_T"],
                             right_on = ["Season", "TeamID"],
                             how = "left")

tournament_result = tournament_result.rename(columns = {"Score" : "LScore"})
tournament_result = tournament_result.drop("TeamID", axis = 1)

print(tournament_result.shape)
tournament_result.head()
```

토너먼트 결과에 토너먼트 시즌별 / 팀별 점수 합계를 합치겠습니다.

```python
tournament_result = pd.merge(tournament_result, tournament_score,
                             left_on = ["Season", "WTeamID_T"],
                             right_on = ["Season", "TeamID"],
                             how = "left")

tournament_result = tournament_result.rename(columns = {"Score_T" : "WScore_T"})
tournament_result = tournament_result.drop("TeamID", axis = 1)

print(tournament_result.shape)
tournament_result.head()
```

```python
tournament_result = pd.merge(tournament_result, tournament_score,
                             left_on = ["Season", "LTeamID_T"],
                             right_on = ["Season", "TeamID"],
                             how = "left")

tournament_result = tournament_result.rename(columns = {"Score_T" : "LScore_T"})
tournament_result = tournament_result.drop("TeamID", axis = 1)

print(tournament_result.shape)
tournament_result.head()
```

```python
tournament_result = tournament_result.rename(columns = {"WTeamID_T" : "WTeamID", "LTeamID_T" : "LTeamID",
                                   "WSeed_num" : "WSeed", "LSeed_num" : "LSeed"})

tournament_result.head()
```

```python
tournament_clean = tournament_result[['WSeed', 'LSeed', 'WScore', 'LScore', 'WScore_T', 'LScore_T']]

print(tournament_clean.shape)
tournament_clean.head()
```

```python
tournament_win = tournament_clean.copy()
tournament_lose = tournament_win.copy()
```

```python
tournament_lose["LSeed"] = tournament_win["WSeed"]
tournament_lose["WSeed"] = tournament_win["LSeed"]

tournament_lose["LScore"] = tournament_win["WScore"]
tournament_lose["WScore"] = tournament_win["LScore"]

tournament_lose["LScore_T"] = tournament_win["WScore_T"]
tournament_lose["WScore_T"] = tournament_win["LScore_T"]

print(tournament_lose.shape)
tournament_lose.head()
```

```python
tournament_win["Seed_diff"] = tournament_win["WSeed"] - tournament_win["LSeed"]
tournament_win["Score_diff"] = tournament_win["WScore"] - tournament_win["LScore"]
tournament_win["Score_T_diff"] = tournament_win["WScore_T"] - tournament_win["LScore_T"]

print(tournament_win.shape)
tournament_win.head()
```

```python
tournament_lose["Seed_diff"] = tournament_lose["WSeed"] - tournament_lose["LSeed"]
tournament_lose["Score_diff"] = tournament_lose["WScore"] - tournament_lose["LScore"]
tournament_lose["Score_T_diff"] = tournament_lose["WScore_T"] - tournament_lose["LScore_T"]

print(tournament_lose.shape)
tournament_lose.head()
```

```python
tournament_win["Result"] = 1
tournament_lose["Result"] = 0

train = pd.concat([tournament_win, tournament_lose])

print(train.shape)
train.head()
```

```python
test = pd.merge(test, MNCAATourneySeeds,
                left_on = ["Season", "WTeamID"],
                right_on = ["Season", "TeamID"],
                how = "left")

test = test.rename(columns = {"Seed" : "WSeed"})
test = test.drop("TeamID", axis = 1)

test = pd.merge(test, MNCAATourneySeeds,
                left_on = ["Season", "LTeamID"],
                right_on = ["Season", "TeamID"],
                how = "left")

test = test.rename(columns = {"Seed" : "LSeed"})
test = test.drop("TeamID", axis = 1)

print(test.shape)
test.head()
```

# Hyperparameter Tuning

## Coarse Search

```python
X_train_kf, X_test_kf, y_train_kf, y_test_kf = \
    train_test_split(X_train, y_train, test_size = 0.3, random_state = 37)


n_estimators = 1000
num_loop = 100
early_stopping_rounds = 20

coarse_hyperparameters_list = []


for loop in range(num_loop):
    learning_rate = np.random.uniform(low = 0.01, high = 0.1)
    max_depth = np.random.randint(low = 2, high = 100)        
    num_leaves = np.random.randint(2, 500)            
    subsample = np.random.uniform(low = 0.1, high = 1.0)
    colsample_bytree = np.random.uniform(low = 0.1, high = 1.0)
    
    
    model = LGBMRegressor(n_estimators = n_estimators,
                          learning_rate = learning_rate,
                          max_depth = max_depth,   
                          num_leaves = num_leaves,                                                    
                          subsample = subsample,                          
                          colsample_bytree = colsample_bytree,
                          random_state = 37)
        
    model.fit(X_train_kf, y_train_kf,
              eval_set = [(X_test_kf, y_test_kf)],    
              verbose = 0,
              early_stopping_rounds = early_stopping_rounds)
    
    y_predict_test_kf = model.predict(X_test_kf)
    
    best_iteration = model.best_iteration_
    score = log_loss(y_test_kf, y_predict_test_kf)

    coarse_hyperparameters_list.append({
        'loop': loop,
        'n_estimators': best_iteration,
        'learning_rate': learning_rate,
        "max_depth" : max_depth,
        'num_leaves': num_leaves,
        'subsample': subsample,
        'colsample_bytree': colsample_bytree,
        'random_state': 37,
        'score': score,
    })

    print(f"{loop:2} best iteration = {best_iteration} Score = {score:.5f}")

coarse_hyperparameters_list = pd.DataFrame(coarse_hyperparameters_list)

coarse_hyperparameters_list = coarse_hyperparameters_list.sort_values(by = "score")

coarse_hyperparameters_list.head()
```

## Finer Search

```python
X_train_kf, X_test_kf, y_train_kf, y_test_kf = \
    train_test_split(X_train, y_train, test_size = 0.3, random_state = 37)

n_estimators = 1000
num_loop = 100
early_stopping_rounds = 20

finer_hyperparameters_list = []

for loop in range(num_loop):
    
    learning_rate = np.random.uniform(low = 0.020531, high = 0.094332)
    max_depth = np.random.randint(low = 50, high = 99)     
    num_leaves = np.random.randint(12, 384)
    subsample = np.random.uniform(low = 0.154675, high = 0.498274)
    colsample_bytree = np.random.uniform(low = 0.691495, high = 0.850703)
    
    model = LGBMRegressor(n_estimators = n_estimators,
                          learning_rate = learning_rate,
                          max_depth = max_depth,   
                          num_leaves = num_leaves,                                                    
                          subsample = subsample,                          
                          colsample_bytree = colsample_bytree,
                          random_state = 37)
        
    model.fit(X_train_kf, y_train_kf,
              eval_set = [(X_test_kf, y_test_kf)],    
              verbose = 0,
              early_stopping_rounds = early_stopping_rounds)
    
    y_predict_test_kf = model.predict(X_test_kf)
    
    best_iteration = model.best_iteration_
    score = log_loss(y_test_kf, y_predict_test_kf)
        
    finer_hyperparameters_list.append({
        'loop': loop,
        'n_estimators': best_iteration,
        'learning_rate': learning_rate,
        'max_depth': max_depth,
        'num_leaves': num_leaves,
        'subsample': subsample,
        'colsample_bytree': colsample_bytree,
        'random_state': 37,
        'score': score,
    })

 
    print(f"{loop:2} best iteration = {best_iteration} Score = {score:.5f}")
    

finer_hyperparameters_list = pd.DataFrame(finer_hyperparameters_list)

finer_hyperparameters_list = finer_hyperparameters_list.sort_values(by = "score")

finer_hyperparameters_list.head()
```
